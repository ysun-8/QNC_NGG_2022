The following code is based on the Poisson distribution tutorial to simulate from photon counting in vision from Sakitt (1972), reproduced in Biophysics: Searching For Principles -- see attached PNG as well for the figure. 

This is the result of an experiment in which observers were asked to rate the intensity of single photons as dim flashes of light, and the resulting distribution looks Poisson!

```python
import numpy as np
from scipy.stats import binom
from scipy.stats import poisson
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output
```

```python
plambda = 0.2
delta_t = 10
max_successes = 7
bins = np.arange(-0.5, max_successes+0.5, 1)
xaxis = (bins[1:] + bins[:-1])/2
num_simulations = 1000

f = plt.figure()
n = 100

p = np.minimum(plambda * delta_t/n, 1)
outcomes = binom.rvs(int(n), p, size=num_simulations)
counts, edges = np.histogram(outcomes, bins)
plt.bar(xaxis, counts/counts.sum())

poissY = poisson.pmf(xaxis, plambda*delta_t)
plt.plot(xaxis, poissY, 'bo-', linewidth=2, markersize=10)

plt.title(f'p={p:.2f}, number bins={n}, number simulations={num_simulations}')
plt.xlabel('Rating')
plt.xticks(np.arange(0, 7, 1))
plt.ylabel('Probability')
plt.legend(['Theoretical Poisson', 'Simulated'], loc='upper right')
plt.show()
```
